```python
# data_extraction.py
import mysql.connector
import pandas as pd

# Connect to MySQL database
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="0028",
    database="customerpurchasedb"
)
```

```python
query = "SELECT * FROM Transactions"
transactions_df = pd.read_sql(query, conn)

conn.close()
```

```python
# Calculate total purchases, total revenue, and average purchase value
total_purchases = transactions_df['purchasequantity'].sum()
total_revenue = (transactions_df['purchasequantity'] * transactions_df['purchaseprice']).sum()
avg_purchase_value = total_revenue / total_purchases
```

```python
# Identify top customers
top_customers = transactions_df.groupby('customerid')['purchasequantity'].sum().sort_values(ascending=False).head(10)
```

```python
# Generate a summary report
summary_report = {
    'Total Purchases': total_purchases,
    'Total Revenue': total_revenue,
    'Average Purchase Value': avg_purchase_value,
    'Top Customers': top_customers, 
}
```
